```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import mglearn

```

```python
digits=load_digits()
fig,axes=plt.subplots(2,5,figsize=(10,5),subplot_kw={'xticks':(),'yticks':()})
for ax, img in zip(axes.ravel(),digits.images):
    ax.imshow(img)
print(digits.keys())  
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(digits.data)
dig_pca=pca.transform(digits.data)
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
 "#A83683", "#4E655E", "#853541", "#3A3120", "#535D8E"]
plt.figure(figsize=(10,10))
plt.xlim(dig_pca[:,0].min(),dig_pca[:,0].max())
plt.ylim(dig_pca[:,1].min(),dig_pca[:,1].max())
for i in range(len(digits.data)):
    plt.text(dig_pca[i,0],dig_pca[i,1],str(digits.target[i]),color=colors[digits.target[i]],fontdict={'weight':'bold','size':9})

    
plt.xlabel("first comp")
plt.ylabel("second comp")
```

```python
from sklearn.manifold import TSNE
tsne=TSNE(random_state=42)
dig_tsne=tsne.fit_transform(digits.data)
plt.figure(figsize=(10,10))
plt.xlim(dig_tsne[:,0].min(),dig_tsne[:,0].max())
plt.ylim(dig_tsne[:,1].min(),dig_tsne[:,1].max())
for i in range(len(digits.data)):
    plt.text(dig_tsne[i,0],dig_tsne[i,1],str(digits.target[i]),color=colors[digits.target[i]],fontdict={'weight':'bold','size':9})

    
plt.xlabel("first comp")
plt.ylabel("second comp")
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

x,y=make_blobs(random_state=1)
km=KMeans(n_clusters=3)
km.fit(x)
```

```python
km.labels_
```

```python
print(km.labels_.shape,x.shape)
```

```python
mglearn.discrete_scatter(x[:,0],x[:,1],km.labels_,markers='o')
mglearn.discrete_scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],[0,1,2],markers='^',markeredgewidth=2)
```

```python
fig,axes=plt.subplots(3,2,figsize=(10,10))
a=axes.ravel()
for i,ax in zip(range(3,9),a):
    km=KMeans(n_clusters=i)
    km.fit(x)
    mglearn.discrete_scatter(x[:,0],x[:,1],km.labels_,ax=ax)
    ax.set_title("{} clusters".format(i))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
    
    
```

```python
from sklearn.datasets import make_moons
x,y=make_moons(n_samples=200,noise=0.05,random_state=0)
km=KMeans(n_clusters=2)
km.fit(x)
ypred=km.predict(x)
plt.scatter(x[:,0],x[:,1],c=ypred,cmap=mglearn.cm2,s=60)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker='^',c=[mglearn.cm2(0),mglearn.cm2(1)],s=100,linewidth=2)
plt.xlabel("f0")
plt.ylabel("f1")
```

```python
mglearn.discrete_scatter(x[:,0],x[:,1],km.labels_)
mglearn.discrete_scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],markers='*',c=[mglearn.cm2(1),mglearn.cm3(2)])
plt.xlabel("F0")
plt.ylabel("F1")
plt.legend(["cluster 1","cluster 2","centers"])
plt.title("KMEANS FOR MAKE_MOONS DATASET")
```

```python
from scipy.cluster.hierarchy import dendrogram,ward
x,y=make_blobs(random_state=0,n_samples=12)
```

```python
linkage_array=ward(x)
dendrogram(linkage_array)
ax=plt.gca()
bounds = ax.get_xbound()

ax.plot(bounds,[7.25,7.25],'--',c='k')
ax.plot(bounds,[4,4],'--',c='k')

ax.text(bounds[1], 7.25, ' two clusters', va='center', fontdict={'size': 15})
ax.text(bounds[1], 4, ' three clusters', va='center', fontdict={'size': 15})
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
```

```python
from sklearn.preprocessing import StandardScaler
x,y=make_moons(n_samples=200,random_state=0,noise=0.05)
scaler=StandardScaler()
scaler.fit(x)
x_s=scaler.transform(x)

dbs=DBSCAN()

clusters=dbs.fit_predict(x_s)
mglearn.discrete_scatter(x_s[:,0],x_s[:,1],clusters)
mglearn.discrete_scatter(x_s[:,0],x_s[:,1],clusters)
plt.xlabel("F0")
plt.ylabel("F1")
import pandas as pd
df=pd.DataFrame({"feature1":x_s[:,1],
               "feature2":x_s[:,0],
                "cluster":[clusters[i] for i in clusters]})
display(df)
```

```python
import numpy as np
from sklearn.metrics.cluster import adjusted_rand_score
x,y=make_moons(n_samples=200,noise=0.05,random_state=0)
scaler=StandardScaler()
scaler.fit(x)
x_s=scaler.transform(x)

fig,axes=plt.subplots(1,4,figsize=(12,8),subplot_kw={'xticks':(),'yticks':()})
algos=[KMeans(n_clusters=2),AgglomerativeClustering(n_clusters=2),DBSCAN()]
random_state=np.random.RandomState(seed=0)
random_clusters=random_state.randint(low=0,high=2,size=len(x))
axes[0].scatter(x_s[:,0],x_s[:,1],c=random_clusters,cmap=mglearn.cm3,s=60)
axes[0].set_title("random assignment-ARI{:.2f}".format(adjusted_rand_score(y,random_clusters)))


for ax,algo in zip(axes[1:],algos):
    clusters=algo.fit_predict(x_s)
    ax.scatter(x_s[:, 0], x_s[:, 1], c=clusters,
 cmap=mglearn.cm3, s=60)
axes[1].set_title("kmeans-ARI  {:.2f}".format(adjusted_rand_score(y,KMeans(n_clusters=2).fit(x_s).predict(x_s))))

axes[2].set_title("agglo-ARI  {:.2f}".format(adjusted_rand_score(y,AgglomerativeClustering(n_clusters=2).fit_predict(x_s))))
axes[3].set_title("dbscan-ARI  {:.2f}".format(adjusted_rand_score(y,DBSCAN().fit_predict(x_s))))
```
